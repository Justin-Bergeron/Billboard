```python
import json
from pprint import pprint

with open('Billboard[3000:3041].json') as f:
    data = json.load(f)

artist_list = []    
song_list = []  


i = 1    
j = 1
while i < 2:
    while j < 10:
        artist = data[i]['0'][j]["Artist"]
        song =  data[i]['0'][j]["Song"]
        j += 1
        artist_list.append(artist)
        song_list.append(song)
    i += 1


print(artist_list)
print(song_list)
```
